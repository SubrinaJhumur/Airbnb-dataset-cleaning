```python
"""deleting redunt columns.
    renaming the columns.
    dropping duplicates.
    cleaning individual column.
    remove the nan values from the dataset.
    check for some more informations."""
```

```python
#is there any missing values present in the dataset
data.isnull().sum()
```

```python
# columns present in the dataset
percent_missing = data.isnull().sum() * 100 / len(data)
```

```python
#see the percent of data missing
percent_missing
```

```python
#some columns to dro
data.drop(columns=['id','reviews per month','review rate number', 'calculated host listings count','availability 365', 'house_rules', 'license'],inplace=True)
```

```python
# is there any missing values present in the dataset
data.isnull().sum()
```

```python
# renaming the column in a way
data.rename(columns={'NAME':'Name'},inplace=True)
```

```python
#first 5 rows of the dataset 
data.head()
```

```python
# finding the duplicated values in the dataset
data.duplicated().sum()
```

```python
# dropping the duplicated values
data.drop_duplicates(inplace=True)
```

```python
# after dropping the duplicated values
data.duplicated().sum()
```

```python
# is there any missing values present in the dataset
data.isna().sum()
```

```python
# drop the last review column from the dataset
data.drop(columns=['last review'],inplace=True)
```

```python
# dropping the null values from the dataset
data=data.dropna()
```

```python
# after dropping the null values from the dataset
data.isna().sum()
```

```python
#first 5 rows of the dataset
data.head()
```

```python
# converting the price column value from str data type to int data type
# first removing the $ sign with empty
data['price']=data['price'].str.replace('$','')
```

```python
# first removing the , sign with empty
data['price']=data['price'].str.replace(',','')
```

```python
#converting the price column value from str data type to int data type
data['price']=data['price'].astype(int)
```

```python
# checking if the column is converting to int type or not
data['price'].info()
```

```python
#first removing the $ sign with empty
data['service fee']=data['service fee'].str.replace('$','')
```

```python
#converting the price column value from str data type to int data type
data['service fee']=data['service fee'].astype(int)
```

```python
#first 5 rows of the dataset
data.head()
```

```python
# check the instant_bookable column value counts figure
data['instant_bookable'].value_counts()
```

```python
# converting instant_bookable columns true or false to 1 and 0
data['instant_bookable'] = data['instant_bookable'].apply(lambda x: 1 if x==True else 0)
```

```python
# check the first 5 rows of the dataset
data.head()
```

```python
data.reset_index(inplace=True)
```

```python
# check the first 5 rows of the dataset
data.head()
```

```python
data.drop(columns=['level_0','index'],inplace=True)
```
